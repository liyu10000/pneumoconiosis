```python
import os
import SimpleITK as sitk
import cv2
import pydicom
import numpy as np
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def loadFile(filename):
    ds = sitk.ReadImage(filename)
    img_array = sitk.GetArrayFromImage(ds)
#     frame_num, width, height = img_array.shape
    return img_array


filename = '/home/sakulaki/yolo-yuli/pneumoconiosis/samples/01.DCM'
img_array = loadFile(filename)
print(type(img_array), len(img_array))
print(img_array.shape)

plt.imshow(img_array[0], cmap = plt.cm.gray)
plt.show()
```

```python
def saveImages(filedir, savedir):
    os.makedirs(savedir, exist_ok=True)
    for file in os.listdir(filedir):
        filename = os.path.join(filedir, file)
        img_array = loadFile(filename)
        imageout = os.path.join(savedir, os.path.splitext(file)[0] + '.jpg')
        Image.fromarray(img_array[0]).save(imageout)
        
filedir = '/home/sakulaki/yolo-yuli/pneumoconiosis/samples'
savedir = '/home/sakulaki/yolo-yuli/pneumoconiosis/sample_jpgs'
saveImages(filedir, savedir)
```
